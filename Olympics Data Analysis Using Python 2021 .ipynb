## Top Countries Participating

```python
top_10_countries=Athletes_Dataset.Team.value_counts().sort_values(ascending=False).head(10)
top_10_countries
```

```python
# Plot for the top 10 countries
plt.figure(figsize=(12,6))
plt.title('Overall Participationby Country')
sns.barplot(x=top_10_countries.index,y=top_10_countries,palette='Set2')
plt.grid(True)
plt.show()
```

```python
# Age Distribution Of the Participation
plt.figure(figsize=(12,6))
plt.title('Age Distribution of the Athletes')
plt.xlabel('Age')
plt.ylabel('Number of Participants')
plt.hist(Athletes_Dataset.Age,bins=np.arange(10,80,2),color='purple',edgecolor='white');
```

## Separate Season in Dataset


```python
# Winter Olympic Sport
winter_sport=Athletes_Dataset[Athletes_Dataset.Season=='Winter'].Sport.unique()
winter_sport
```

```python
# Summer Olympic Sport
summer_sport=Athletes_Dataset[Athletes_Dataset.Season=='Summer'].Sport.unique()
winter_sport
```

```python
# male and female Participants
gender_count=Athletes_Dataset['Sex'].value_counts()
gender_count
```

```python
# Pie plot for male and female athletes
plt.figure(figsize=(12,6))
plt.title('Gender Distribution of the Athletes')
plt.pie(gender_count,labels=gender_count.index,autopct='%1.1f%%',startangle=150,shadow=True);
```

```python
# Total Medals
Athletes_Dataset['Medal'].value_counts()
```

```python
women_olympics=Athletes_Dataset[(Athletes_Dataset.Sex=='F') & (Athletes_Dataset.Season=='Summer')]
women_olympics.head()
```

```python
# CountPlot
sns.set(style="darkgrid")
plt.figure(figsize=(12,6))
sns.countplot(x='Year',data=women_olympics,palette="Spectral")
plt.title('Women Participation')
plt.grid(True)
plt.show()
```

```python
part=women_olympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(12,6))
part.loc[:,'F'].plot()
plt.title('Plot of Female Athletes over time ')
plt.grid(True)
plt.show()
```

```python
# GOLD Medal athletes
gold_medals=Athletes_Dataset[(Athletes_Dataset.Medal=='Gold')]
gold_medals.head()
```

```python
# Take only the values that are different from NaN
goldmedals=gold_medals[np.isfinite(gold_medals['Age'])]
goldmedals.head()
```

```python
# Gold beyond 60
goldmedals['ID'][goldmedals['Age']>60].count()
```

```python
sporting_events=goldmedals['Sport'][goldmedals['Age']>60]
sporting_events
```

```python
# Plot for sporting_events
plt.figure(figsize=(12,6))
plt.tight_layout()
sns.countplot(sporting_events)
plt.title('Gold Medal for Athletes over 60 Year ')
plt.grid(True)
plt.show()
```

```python
# Gold Medal for Each Country
goldmedals.Region.value_counts().reset_index(name='Medal')
```

```python
Total_Gold_Medal=goldmedals.Region.value_counts().reset_index(name='Medal').head(5)
g=sns.catplot(x="index",y="Medal",data=Total_Gold_Medal,height=5,kind="bar",palette="rocket")
g.despine(left=True)
g.set_xlabels("Top 5 Countries")
g.set_ylabels("Number of Medals")
plt.title("Gold Medal per Country")
plt.grid(True)
plt.show()
```

## Rio Olympics
